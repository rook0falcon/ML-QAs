```
import os
```

```
sorted(os.listdir('/content/A/'))
```

```
import pandas as pd
```

```
Act_code = pd.DataFrame({'Code':['Walking','Jogging','Stairs','Sitting','Standing','Typing','Brushing teeth','Eating Soup','Eating Chips',
 'Eating Pasta','Drinking from Cup','Eating Sandwich','Kicking','Playing Catch','Dribbling','Writing','Clapping','Folding cloth'], 
 'Activity':['A','B','C','D','E','F','G','H','I','J','K','L','M','O','P','Q','R','S']})
Act_code
```

```
%%time
data = []
for i in sorted(os.listdir('/content/A/')):
    with open('/content/A/'+i, 'r') as f:
        text = f.read()
        temp = text.split(';\n')
        final = []
        for i in range (len(temp)):
            a = temp[i].split(',')
            final.append(a)
        data += final
print(len(data))
```

```
data[:10]
```

```
data = pd.DataFrame(data, columns = ['Subject-id', 'Activity Label', 'Timestamp', 'x', 'y', 'z'])
data.head(10)
```

```
(data.isnull()).sum()
```

```
data = data.dropna()
```

```
(data.isnull()).sum()
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
target = le.fit_transform(data['Activity Label'])
```

```
%%time
# convert the elements of the dataframe from string to numeric
train = data.drop(columns = 'Activity Label', axis = 1)
train = train.apply(pd.to_numeric, errors='coerce')
```

```
from datetime import datetime
```

```
%%time
train['Timestamp'] = train['Timestamp'].apply(lambda x: x//1000000)
train['Timestamp'] = train['Timestamp'].apply(lambda x: datetime.fromtimestamp(x))
```

```
train.head()
```

```
for time in ('year','month','week','day','hour','minute','second'):
    train[time] = getattr(train['Timestamp'].dt,time)
train.drop(columns=["Timestamp","Subject-id"],inplace=True)
```

```
train.head()
```

```
len(train)
```

```
train['Activity Label'] = target
```

```
train = train.sample(frac=0.05).reset_index(drop=True)
len(train)
```

```
train.head()
```

```
train['Activity Label'].value_counts()
```

```
from sklearn.model_selection import train_test_split
```

```
train.shape
```

```
y=train["Activity Label"]
x=train.drop(columns="Activity Label")
x_train, x_test, y_train, y_test = train_test_split(x,y , train_size = 0.8, random_state =  42, stratify = y)
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train , train_size = 0.8, random_state =  42, stratify = y_train)
```

```
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape, y_test.shape)
```

```
x_train.head()
```

```
y_train.head()
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```
%%time
model = RandomForestClassifier(n_estimators=100, max_depth=10)
model.fit(x_train, y_train)
```

```
y_pred= model.predict(x_test)
acc = accuracy_score(y_test, y_pred)
```
